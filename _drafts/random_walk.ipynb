```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

plt.rcParams["animation.html"] = "jshtml"
plt.rcParams['figure.dpi'] = 150 

L = 10.0  # length of the rod
Nx = 51  # number of grid points
dx = L / (Nx - 1)  # spacing between grid points
alpha = 0.01  # thermal diffusivity
Nt = 1000  # number of time steps
dt = 0.1  # time step duration
c = alpha * dt / dx**2

T = np.zeros(Nx)
T[0], T[-1] = 0, 100  # Boundary conditions


T_anim = [T.copy()]
for _ in range(Nt):
    T[1:-1] = T[1:-1] + c * (T[2:] - 2*T[1:-1] + T[:-2])
    if _ % 100 == 0:
        T_anim.append(T.copy())

fig, ax = plt.subplots(figsize=(10, 4))
line, = ax.plot(np.linspace(0, L, Nx), T_anim[0], color='blue')
ax.set_xlim(0, L)
ax.set_ylim(0, 110)
ax.set_title('Temperature Distribution Over Time')
ax.set_xlabel('Position along the rod')
ax.set_ylabel('Temperature')

def update(frame):
    line.set_ydata(T_anim[frame])
    return line,

anim = FuncAnimation(fig, update, frames=len(T_anim), interval=100, blit=True)
plt.close()  # Prevents duplicate display
anim
```
