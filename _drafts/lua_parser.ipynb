```python
%%script node
function parseLuaTable(luaString) {
  // Remove any whitespace and newlines
  luaString = luaString.replace(/\s/g, '');
  
  // Function to parse a single value
  function parseValue() {
    if (luaString[0] === '{') {
      return parseTable();
    } else if (luaString[0] === '"' || luaString[0] === "'") {
      return parseString();
    } else if (!isNaN(luaString[0]) || luaString[0] === '-') {
      return parseNumber();
    } else {
      throw new Error('Unexpected token: ' + luaString[0]);
    }
  }
  
  // Function to parse a table
  function parseTable() {
    let result = {};
    luaString = luaString.slice(1); // Remove opening '{'
    
    while (luaString[0] !== '}') {
      let key;
      if (luaString[0] === '[') {
        luaString = luaString.slice(1);
        key = parseValue();
        luaString = luaString.slice(1); // Remove closing ']'
      } else {
        key = parseString();
      }
      
      luaString = luaString.slice(1); // Remove '='
      let value = parseValue();
      result[key] = value;
      
      if (luaString[0] === ',') {
        luaString = luaString.slice(1);
      }
    }
    
    luaString = luaString.slice(1); // Remove closing '}'
    return result;
  }
  
  // Function to parse a string
  function parseString() {
    let endQuote = luaString[0];
    let endIndex = luaString.indexOf(endQuote, 1);
    let result = luaString.slice(1, endIndex);
    luaString = luaString.slice(endIndex + 1);
    return result;
  }
  
  // Function to parse a number
  function parseNumber() {
    let endIndex = luaString.search(/[,}]/);
    let result = parseFloat(luaString.slice(0, endIndex));
    luaString = luaString.slice(endIndex);
    return result;
  }
  
  return parseValue();
}

// Example usage
let luaData = `{
  name = "John",
  age = 30,
  ["key with spaces"] = "value",
  nested = {
    x = 10,
    y = 20
  },
  [1] = "first",
  [2] = "second"
}`;

let result = parseLuaTable(luaData);
console.log(JSON.stringify(result, null, 2));
```
